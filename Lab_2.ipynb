```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load image as grayscale
image = cv2.imread('grayscale_image.jpg', cv2.IMREAD_GRAYSCALE)

# Perform 2D DFT
f = np.fft.fft2(image)
fshift = np.fft.fftshift(f)  # Shift zero frequency components to the center
magnitude_spectrum = np.abs(fshift)

# Plot the magnitude spectrum
plt.imshow(np.log(magnitude_spectrum + 1), cmap='gray')
plt.title('Magnitude Spectrum')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Flatten the image
image_flatten = image.flatten().reshape(1, -1)

# Apply PCA for KL Transform
pca = PCA(n_components=2)
image_klt = pca.fit_transform(image_flatten.T)

# Plot the first two components of the transformed image
plt.plot(image_klt[:, 0], image_klt[:, 1])
plt.show()
```
